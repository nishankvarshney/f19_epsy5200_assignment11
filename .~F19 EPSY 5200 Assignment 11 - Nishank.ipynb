```python
#3a)
sns.pairplot(school_data_2017[['Average_Class_Size', 'Average_Salary', 'Perc_Economically_Disadvantaged', 'Perc_English_Language_Learner']], kind = "reg")
```

```python
#3b)
# We see that the Average Salary increases with an increase in class size and percentage of english language learners, but decreases with percentage of economically disadvantaged.
# I would have expected the pairplot of same variables on X and Y to be a linear plot, but instead it gave histograms.
```

```python
#3c)
# outcome = Average_Salary
# predictor = Perc_Economically_Disadvantaged
reg_income_ed = smf.ols(formula = 'Average_Salary ~ Perc_Economically_Disadvantaged', data = school_data_2017).fit()
reg_income_ed.summary()
```

```python
# The regression results show that a unit increase in percentage of economically disadvantaged children is associated with
# a nearly 30 unit decrease in average salary, and the results are statistically significant with a p-value of 0.000
```
